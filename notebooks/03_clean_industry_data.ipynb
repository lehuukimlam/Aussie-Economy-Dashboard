```python
# Function to extract Industry Name and update Indicator Name for df_employed_melted
def process_employed_indicator(indicator_name):
    parts = indicator_name.split(' ; ')
    if len(parts) > 1:
        industry_name = parts[0]
        indicator = parts[1].strip()
        return industry_name, indicator
    # If the split doesn't result in more than one part, the original name is likely the industry name
    return indicator_name, ""


# Apply the function to create 'Industry Name' and update 'Indicator Name' in df_employed_melted
df_employed_melted[['Industry Name', 'Indicator Name']] = df_employed_melted['Indicator Name'].apply(process_employed_indicator).tolist()

# Function to extract Industry Name and update Indicator Name for df_gdp_melted
def process_gdp_indicator(indicator_name):
    parts = indicator_name.split(' ; ')
    if len(parts) > 1:
        industry_name = parts[0].split(' (')[0].strip() # Extract industry name before ( and strip
        indicator = ' ; '.join(parts[1:]).strip()
        # If indicator is empty, it means the original name was just the industry
        if not indicator:
            indicator = industry_name
        return industry_name, indicator
    return indicator_name, indicator_name


# Apply the function to create 'Industry Name' and update 'Indicator Name' in df_gdp_melted
df_gdp_melted[['Industry Name', 'Indicator Name']] = df_gdp_melted['Indicator Name'].apply(process_gdp_indicator).tolist()


# Display the head of the modified DataFrames
print("Modified DataFrame (Employed By Industry):")
display(df_employed_melted.head())

print("\nModified DataFrame (GDP by Industry):")
display(df_gdp_melted.head())
```

```python
# Clean 'Indicator Name' and 'Industry Name' columns in df_employed_melted
df_employed_melted['Indicator Name'] = df_employed_melted['Indicator Name'].str.replace(';', '').str.replace('.', '', regex=False).str.replace('\d+', '', regex=True).str.strip()
df_employed_melted['Industry Name'] = df_employed_melted['Industry Name'].str.replace(';', '').str.replace('.', '', regex=False).str.replace('\d+', '', regex=True).str.strip()


# Clean 'Indicator Name' and 'Industry Name' columns in df_gdp_melted
df_gdp_melted['Indicator Name'] = df_gdp_melted['Indicator Name'].str.replace(';', '').str.replace('.', '', regex=False).str.replace('\d+', '', regex=True).str.strip()
df_gdp_melted['Industry Name'] = df_gdp_melted['Industry Name'].str.replace(';', '').str.replace('.', '', regex=False).str.replace('\d+', '', regex=True).str.strip()


# Display the head of the cleaned DataFrames
print("Cleaned DataFrame (Employed By Industry):")
display(df_employed_melted.head())

print("\nCleaned DataFrame (GDP by Industry):")
display(df_gdp_melted.head())
```
