```python
import pandas as pd

excel_file_gdp = "/content/Raw GDP By States.xlsx"
df_gdp = pd.read_excel(excel_file_gdp)
# Delete rows from Unit to Collection Month (index 0-4)
df_gdp = df_gdp.drop(df_gdp.index[0:9]).copy()
display(df_gdp.head())
```

```python
excel_file_unem = "/content/Raw Unem by States.xlsx"
df_unem = pd.read_excel(excel_file_unem)
# Delete rows from Unit to Collection Month (index 0-4)
df_unem = df_unem.drop(df_unem.index[0:9]).copy()
display(df_unem.head())
```

```python
excel_file_cpi = "/content/Raw CPI by States.xlsx"
df_cpi = pd.read_excel(excel_file_cpi)
# Delete rows from Unit to Collection Month (index 0-4)
df_cpi = df_cpi.drop(df_cpi.index[0:9]).copy()
display(df_cpi.head())
```

```python
# Transpose the DataFrame, using the first row as new headers
df_cpi_transposed = df_cpi.transpose().reset_index()

# Rename columns using the values from the first row
df_cpi_transposed.columns = df_cpi_transposed.iloc[0]
df_cpi_transposed = df_cpi_transposed.drop(0)

# Rename the first column to 'Indicator name'
df_cpi_transposed = df_cpi_transposed.rename(columns={'Unnamed: 0': 'Indicator name'})

# Extract City name from 'Indicator name'
df_cpi_transposed['City Name'] = df_cpi_transposed['Indicator name'].apply(lambda x: x.split(';')[-2].strip() if isinstance(x, str) and len(x.split(';')) > 2 else 'Australia' if isinstance(x, str) and 'Australia' in x else None)

# Melt the DataFrame to long format
df_cpi_normalized = df_cpi_transposed.melt(id_vars=['Indicator name', 'City Name'], var_name='Date', value_name='Value')

# Add a 'State name' column with None values
df_cpi_normalized['State name'] = None

# Keep only relevant columns and reorder
df_cpi_normalized = df_cpi_normalized[['Indicator name', 'State name', 'City Name', 'Date', 'Value']]

display(df_cpi_normalized.head())
```

```python
# Transpose the DataFrame, using the first row as new headers
df_unem_transposed = df_unem.transpose().reset_index()

# Rename columns using the values from the first row
df_unem_transposed.columns = df_unem_transposed.iloc[0]
df_unem_transposed = df_unem_transposed.drop(0)

# Rename the first column to 'Indicator name'
df_unem_transposed = df_unem_transposed.rename(columns={'Unnamed: 0': 'Indicator name'})

# Extract State name from 'Indicator name'
df_unem_transposed['State name'] = df_unem_transposed['Indicator name'].apply(lambda x: x.split('>')[-1].split(';')[0].strip() if isinstance(x, str) and '>' in x else 'Australia' if isinstance(x, str) and 'Australia' in x else None)

# Melt the DataFrame to long format
df_unem_normalized = df_unem_transposed.melt(id_vars=['Indicator name', 'State name'], var_name='Date', value_name='Value')

# Add a 'City Name' column with None values
df_unem_normalized['City Name'] = None

# Keep only relevant columns and reorder
df_unem_normalized = df_unem_normalized[['Indicator name', 'State name', 'City Name', 'Date', 'Value']]

display(df_unem_normalized.head())
```

```python
# Transpose the DataFrame, using the first row as new headers
df_gdp_transposed = df_gdp.transpose().reset_index()

# Rename columns using the values from the first row
df_gdp_transposed.columns = df_gdp_transposed.iloc[0]
df_gdp_transposed = df_gdp_transposed.drop(0)

# Rename the first column to 'Indicator name'
df_gdp_transposed = df_gdp_transposed.rename(columns={'Unnamed: 0': 'Indicator name'})

# Extract State name from 'Indicator name'
df_gdp_transposed['State name'] = df_gdp_transposed['Indicator name'].apply(lambda x: x.split(';')[0].strip() if ';' in str(x) else None)

# Melt the DataFrame to long format
df_gdp_normalized = df_gdp_transposed.melt(id_vars=['Indicator name', 'State name'], var_name='Date', value_name='Value')

# Add a 'City Name' column with None values
df_gdp_normalized['City Name'] = None

# Keep only relevant columns and reorder
df_gdp_normalized = df_gdp_normalized[['Indicator name', 'State name', 'City Name', 'Date', 'Value']]

display(df_gdp_normalized.head())
```

```python
# Combine the normalized dataframes
combined_df = pd.concat([df_cpi_normalized, df_unem_normalized, df_gdp_normalized], ignore_index=True)

display(combined_df.head())
```

```python
# Export the combined DataFrame to a CSV file
combined_df.to_csv('combined_data.csv', index=False)

print("combined_df exported to combined_data.csv")
```
